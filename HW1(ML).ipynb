```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('MEDV', axis =1),
                                                    df['MEDV'], 
                                                    test_size=0.22, 
                                                    random_state=1888)
```

```python
model_lr = LinearRegression()
```

```python
model_lr.fit(X_train,y_train)
```

```python
pred = model_lr.predict(X_test)
```

```python
print('Ошибка MSE=',mean_squared_error(y_test,pred,squared=True))
```

```python
plt.figure(figsize=(20,9))
plt.plot(y_test.values, "blue", linewidth=5, label = 'Реальные значения')
plt.plot(pred, "red", linewidth=5, label = 'Наш первый ML')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
